```python
for directory in glob.glob(path_to_data):
    data = pd.read_excel(os.path.join(directory, 'data.xls'))
    meta = pd.read_excel(os.path.join(directory, 'meta.xls'))
    # break
```

```python
data.loc[data['Описание'].isna(), 'Описание'] = 'NONE'
data = data.loc[~data['Цена'].isna()]
# data['Цена'] = data['Цена'].str.replace('pуб.', '').str.split().str.join('')
```

```python
data = pd.read_excel('./parsing_pro-bike.xls')
data.head()
```

```python
cols = data.columns[:5]
cols
```

```python
data = data[cols]
data.head()
```

```python
data.dropna(inplace=True)
```

```python
data['Цена'] = data['Цена'].apply(lambda x: int(''.join(x[:-5].split())))
```

```python
es = Elasticsearch([{'host':'localhost','port':9200}])
```

```python
for row in tqdm.tqdm(data.iterrows(), total=len(data)):
    row = row[1]
    body = {'cat': row['Категория'],
            'item': row['Название'],
            'desc': row['Описание'],
            'price': int(row['Цена']),
            'cb': int(row['CASH_BACK_HEIGHT'])}
    es.index(index='offer', doc_type='sport', body=body)
```
